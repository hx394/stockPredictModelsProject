```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
BASE_DIR = os.path.abspath(os.path.join("..", "..","hongzhen xu","models and results"))  # Get data from csv files in models and results folder
default_AAPL = pd.read_csv(BASE_DIR+'/default/AAPL/actuals_vs_predictions.csv')
default_IBM = pd.read_csv(BASE_DIR+'/default/IBM/actuals_vs_predictions.csv')
default_META = pd.read_csv(BASE_DIR+'/default/META/actuals_vs_predictions.csv')
default_MSFT = pd.read_csv(BASE_DIR+'/default/MSFT/actuals_vs_predictions.csv')
default_NVDA = pd.read_csv(BASE_DIR+'/default/NVDA/actuals_vs_predictions.csv')
default_TSLA = pd.read_csv(BASE_DIR+'/default/TSLA/actuals_vs_predictions.csv')
default_VZ = pd.read_csv(BASE_DIR+'/default/VZ/actuals_vs_predictions.csv')
```

```python
# Extract Data
stockList = ["AAPL","IBM","META","MSFT","NVDA","TSLA","VZ"]
dataList = [default_AAPL,default_IBM,default_META,default_MSFT,default_NVDA,default_TSLA,default_VZ]
# default_actual = default_AAPL['Actual Price']
# default_predicted = default_AAPL['Predicted Price']

for (index,stock) in enumerate(stockList):
    default_actual = dataList[index]['Actual Price']
    default_predicted = dataList[index]['Predicted Price']
    # Plot Predicted Price using default model
    plt.figure(figsize=(14, 8))
    plt.plot(default_actual, label="Actual Price (default)", color='blue', alpha=0.6)
    plt.plot(default_predicted, label="Predicted Price (default)", color='red', linestyle='--', alpha=0.7)
    plt.title("default Prediction Comparison on stock " + str(stock))
    plt.xlabel("Index")
    plt.ylabel("Price")
    plt.legend()
    plt.grid()
    plt.show()

    # Calculate Performance Index
    default_mse = mean_squared_error(default_actual, default_predicted)
    default_mae = mean_absolute_error(default_actual, default_predicted)
    default_r2 = r2_score(default_actual, default_predicted)

    default_metrics = {'MSE': default_mse, 'MAE': default_mae, 'R^2': default_r2}


    # gru_metrics, lstm_metrics

    gru_lstm_metrics_table = pd.DataFrame([default_metrics], index=['default'])
    print(gru_lstm_metrics_table)
```
