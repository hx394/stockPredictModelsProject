```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
BASE_DIR = os.path.abspath(os.path.join("..", "..","hongzhen xu","models and results"))  # Get data from csv files in models and results folder

GRU_AAPL = pd.read_csv(BASE_DIR+'/GRU rate/AAPL/actuals_vs_predictions.csv')
LSTM_AAPL = pd.read_csv(BASE_DIR+'/LSTM rate/AAPL/actuals_vs_predictions.csv')
GRU_IBM = pd.read_csv(BASE_DIR+'/GRU rate/IBM/actuals_vs_predictions.csv')
LSTM_IBM = pd.read_csv(BASE_DIR+'/LSTM rate/IBM/actuals_vs_predictions.csv')
GRU_META = pd.read_csv(BASE_DIR+'/GRU rate/META/actuals_vs_predictions.csv')
LSTM_META = pd.read_csv(BASE_DIR+'/LSTM rate/META/actuals_vs_predictions.csv')
GRU_MSFT = pd.read_csv(BASE_DIR+'/GRU rate/MSFT/actuals_vs_predictions.csv')
LSTM_MSFT = pd.read_csv(BASE_DIR+'/LSTM rate/MSFT/actuals_vs_predictions.csv')
GRU_NVDA = pd.read_csv(BASE_DIR+'/GRU rate/NVDA/actuals_vs_predictions.csv')
LSTM_NVDA = pd.read_csv(BASE_DIR+'/LSTM rate/NVDA/actuals_vs_predictions.csv')
GRU_TSLA = pd.read_csv(BASE_DIR+'/GRU rate/TSLA/actuals_vs_predictions.csv')
LSTM_TSLA = pd.read_csv(BASE_DIR+'/LSTM rate/TSLA/actuals_vs_predictions.csv')
GRU_VZ = pd.read_csv(BASE_DIR+'/GRU rate/VZ/actuals_vs_predictions.csv')
LSTM_VZ = pd.read_csv(BASE_DIR+'/LSTM rate/VZ/actuals_vs_predictions.csv')

default_AAPL = pd.read_csv(BASE_DIR+'/default/AAPL/actuals_vs_predictions.csv')
default_IBM = pd.read_csv(BASE_DIR+'/default/IBM/actuals_vs_predictions.csv')
default_META = pd.read_csv(BASE_DIR+'/default/META/actuals_vs_predictions.csv')
default_MSFT = pd.read_csv(BASE_DIR+'/default/MSFT/actuals_vs_predictions.csv')
default_NVDA = pd.read_csv(BASE_DIR+'/default/NVDA/actuals_vs_predictions.csv')
default_TSLA = pd.read_csv(BASE_DIR+'/default/TSLA/actuals_vs_predictions.csv')
default_VZ = pd.read_csv(BASE_DIR+'/default/VZ/actuals_vs_predictions.csv')
```

```python
# Extract Data
gru_actual_AAPL = GRU_AAPL['Actual Price']
gru_predicted_AAPL = GRU_AAPL['Predicted Price']
lstm_actual_AAPL = LSTM_AAPL['Actual Price']
lstm_predicted_AAPL = LSTM_AAPL['Predicted Price']
# default_actual_AAPL = default_AAPL['Actual Price']
# default_predicted_AAPL = default_AAPL['Predicted Price']

# Plot the prediction comparison of GRU and LSTM models
plt.figure(figsize=(14, 8))
plt.plot(gru_actual_AAPL, label="Actual Price", color='blue', alpha=0.6)
plt.plot(gru_predicted_AAPL, label="Predicted Price (GRU)", color='red', linestyle='--', alpha=0.7)
plt.plot(lstm_predicted_AAPL, label="Predicted Price (LSTM)", color='green', linestyle='--', alpha=0.7)
# plt.plot(default_predicted_AAPL, label="Predicted Price (default)", color='black', linestyle='--', alpha=0.7)
plt.title("GRU, LSTM Prediction Comparison on stock AAPL")
plt.xlabel("Index")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()

# Calculate Performance Index
gru_mse = mean_squared_error(gru_actual_AAPL, gru_predicted_AAPL)
gru_mae = mean_absolute_error(gru_actual_AAPL, gru_predicted_AAPL)
gru_r2 = r2_score(gru_actual_AAPL, gru_predicted_AAPL)

lstm_mse = mean_squared_error(lstm_actual_AAPL, lstm_predicted_AAPL)
lstm_mae = mean_absolute_error(lstm_actual_AAPL, lstm_predicted_AAPL)
lstm_r2 = r2_score(lstm_actual_AAPL, lstm_predicted_AAPL)

# default_mse = mean_squared_error(default_actual_AAPL, default_predicted_AAPL)
# default_mae = mean_absolute_error(default_actual_AAPL, default_predicted_AAPL)
# default_r2 = r2_score(default_actual_AAPL, default_predicted_AAPL)

gru_metrics = {'MSE': gru_mse, 'MAE': gru_mae, 'R2': gru_r2}
lstm_metrics = {'MSE': lstm_mse, 'MAE': lstm_mae, 'R2': lstm_r2}
# default_metrics = {'MSE': default_mse, 'MAE': default_mae, 'R2': default_r2}

# gru_metrics, lstm_metrics

# gru_lstm_metrics_table = pd.DataFrame([gru_metrics, lstm_metrics, default_metrics], index=['GRU', 'LSTM', 'default'])
gru_lstm_metrics_table = pd.DataFrame([gru_metrics, lstm_metrics], index=['GRU', 'LSTM'])
gru_lstm_metrics_table
```
