```python
import pandas as pd
GRU_AAPL = pd.read_csv('GRU rate/AAPL/actuals_vs_predictions.csv')
LSTM_AAPL = pd.read_csv('LSTM rate/AAPL/actuals_vs_predictions.csv')
GRU_IBM = pd.read_csv('GRU rate/IBM/actuals_vs_predictions.csv')
LSTM_IBM = pd.read_csv('LSTM rate/IBM/actuals_vs_predictions.csv')
GRU_META = pd.read_csv('GRU rate/META/actuals_vs_predictions.csv')
LSTM_META = pd.read_csv('LSTM rate/META/actuals_vs_predictions.csv')
GRU_MSFT = pd.read_csv('GRU rate/MSFT/actuals_vs_predictions.csv')
LSTM_MSFT = pd.read_csv('LSTM rate/MSFT/actuals_vs_predictions.csv')
GRU_NVDA = pd.read_csv('GRU rate/NVDA/actuals_vs_predictions.csv')
LSTM_NVDA = pd.read_csv('LSTM rate/NVDA/actuals_vs_predictions.csv')
GRU_TSLA = pd.read_csv('GRU rate/TSLA/actuals_vs_predictions.csv')
LSTM_TSLA = pd.read_csv('LSTM rate/TSLA/actuals_vs_predictions.csv')
GRU_VZ = pd.read_csv('GRU rate/VZ/actuals_vs_predictions.csv')
LSTM_VZ = pd.read_csv('LSTM rate/VZ/actuals_vs_predictions.csv')

standard_AAPL = pd.read_csv('Standard/AAPL/actuals_vs_predictions.csv')
standard_IBM = pd.read_csv('Standard/IBM/actuals_vs_predictions.csv')
standard_META = pd.read_csv('Standard/META/actuals_vs_predictions.csv')
standard_MSFT = pd.read_csv('Standard/MSFT/actuals_vs_predictions.csv')
standard_NVDA = pd.read_csv('Standard/NVDA/actuals_vs_predictions.csv')
standard_TSLA = pd.read_csv('Standard/TSLA/actuals_vs_predictions.csv')
standard_VZ = pd.read_csv('Standard/VZ/actuals_vs_predictions.csv')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 提取数据
gru_actual_AAPL = GRU_AAPL['Actual Price']
gru_predicted_AAPL = GRU_AAPL['Predicted Price']
lstm_actual_AAPL = LSTM_AAPL['Actual Price']
lstm_predicted_AAPL = LSTM_AAPL['Predicted Price']
standard_actual_AAPL = standard_AAPL['Actual Price']
standard_predicted_AAPL = standard_AAPL['Predicted Price']

# 绘制 GRU 和 LSTM 模型的预测对比图
plt.figure(figsize=(14, 8))
plt.plot(gru_actual_AAPL, label="Actual Price (GRU)", color='blue', alpha=0.6)
plt.plot(gru_predicted_AAPL, label="Predicted Price (GRU)", color='red', linestyle='--', alpha=0.7)
plt.plot(lstm_predicted_AAPL, label="Predicted Price (LSTM)", color='green', linestyle='--', alpha=0.7)
plt.plot(standard_predicted_AAPL, label="Predicted Price (Standard)", color='black', linestyle='--', alpha=0.7)
plt.title("GRU, LSTM, standard Prediction Comparison on stock AAPL")
plt.xlabel("Index")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()

# 计算性能指标
gru_mse = mean_squared_error(gru_actual_AAPL, gru_predicted_AAPL)
gru_mae = mean_absolute_error(gru_actual_AAPL, gru_predicted_AAPL)
gru_r2 = r2_score(gru_actual_AAPL, gru_predicted_AAPL)

lstm_mse = mean_squared_error(lstm_actual_AAPL, lstm_predicted_AAPL)
lstm_mae = mean_absolute_error(lstm_actual_AAPL, lstm_predicted_AAPL)
lstm_r2 = r2_score(lstm_actual_AAPL, lstm_predicted_AAPL)

standard_mse = mean_squared_error(standard_actual_AAPL, standard_predicted_AAPL)
standard_mae = mean_absolute_error(standard_actual_AAPL, standard_predicted_AAPL)
standard_r2 = r2_score(standard_actual_AAPL, standard_predicted_AAPL)

gru_metrics = {'MSE': gru_mse, 'MAE': gru_mae, 'R2': gru_r2}
lstm_metrics = {'MSE': lstm_mse, 'MAE': lstm_mae, 'R2': lstm_r2}
standard_metrics = {'MSE': standard_mse, 'MAE': standard_mae, 'R2': standard_r2}

# gru_metrics, lstm_metrics

gru_lstm_metrics_table = pd.DataFrame([gru_metrics, lstm_metrics, standard_metrics], index=['GRU', 'LSTM', 'Standard'])
gru_lstm_metrics_table
```
